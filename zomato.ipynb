```python
data1=pd.read_excel('Country-Code.xlsx')
data1.head()
```

```python
data1.columns
```

```python
final_data=pd.merge(data,data1,on='Country Code', how='left')
```

```python
final_data.head()
```

```python
final_data.dtypes
```

```python
final_data.columns
```

```python
country_names=final_data.Country.value_counts().index
```

```python
country_val=final_data.Country.value_counts().values
```

```python
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
plt.show()
```

```python
ratings=final_data.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings.head()
```

```python
ratings
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
plt.show()
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
plt.show()
```

```python
sns.countplot(x="Rating color",data=ratings,palette=['blue','red','orange','yellow','green','green'])
plt.show()
```
